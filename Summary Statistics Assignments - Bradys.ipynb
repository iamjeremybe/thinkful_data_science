```python
import pandas as pd
import numpy as np
from scipy import stats
import statistics
```

```python
# The ages of the original Brady children
bradys = np.array([14, 12, 11, 10, 8, 6, 8])

print(bradys)
#print(type(bradys))

# Exercise 1:
# Mean, median, mode--by hand
mean_brady_age = sum(bradys) / len(bradys)
print(mean_brady_age)

print(np.sort(bradys))
median_brady_age = 10 # I eyeballed it.

mode_brady_age = 8 # I eyeballed it.

# Variance is sum of (difference between each value and mean, squared), divided by sample size 
# NOTE: (I originally subtracted 1, not considering that this is the entire population of Bradys and not just a sample. :)

# Standard deviation is the sqrt of variance
variance_bradys = sum(map(lambda x: (x - mean_brady_age)**2, bradys)) / (len(bradys)) # - 1) <--THIS IS NOT A SAMPLE.
std_dev_bradys = np.sqrt(variance_bradys)
print("Variance: {}\nStandard deviation: {}".format(variance_bradys,std_dev_bradys))

# Standard error is standard deviation divided by sqrt of sample size
std_err_bradys = std_dev_bradys / np.sqrt(len(bradys) - 1)
print ("Standard error: {}".format(std_err_bradys))

# Now run the functions to calculate these values, and see how your work compares

df_bradys = pd.DataFrame()
df_bradys['ages'] = bradys
np.median(bradys)
print("Median: {}\nMode: {}".format(statistics.median(bradys),stats.mode(bradys)[0][0]))
df_bradys.describe()
```

### Exercise 5:
    TV Guide 20% fans 
    Entertainment Weekly 23% fans 
    Pop Culture Today 17% fans 
    SciPhi Phanatic 5% fans

Based on these numbers, what percentage of adult Americans would you estimate were Brady Bunch fans on the 50th anniversary of the show?


* I'm interested to know how representative of "adult Americans" we consider polls of readers of four media- or entertainment-oriented magazines. Is the selection of magazines/subscribers already introducing bias? 
* I mean, I can run the numbers, but without a more general, random sampling of adult Americans, it's possible (probable?) that the results aren't truly representative. For example, it's likely that Sports Illustrated subscribers believe an event like the World Series or the Super Bowl carries a lot more significance than subscribers to, say, SciPhi Phanatic. :)
* Speaking of SciPhi Phanatic, they are a clear outlier here--and maybe their lower result provides balance against the probable Brady-friendly bias of the other three sources? This poll result is going to bump up the variance and standard deviation though, making it harder to estimate the actual percentage of Brady Bunch fans among American adults.

```python
brady_fans = np.array([20, 23, 17, 5])
df_brady_fans = pd.DataFrame()
df_brady_fans['polls'] = brady_fans

mean_brady_fans = sum(brady_fans) / len(brady_fans)
print("Median: {}".format(np.median(brady_fans)))

variance_brady_fans = sum(map(lambda x: (x - mean_brady_fans)**2, brady_fans)) / (len(brady_fans) - 1)
std_dev_brady_fans = np.sqrt(variance_brady_fans)
print("Variance: {}\nStandard deviation: {}".format(variance_brady_fans,std_dev_brady_fans))

# Standard error is standard deviation divided by sqrt of sample size
std_err_brady_fans = std_dev_brady_fans / np.sqrt(len(brady_fans) - 1)
print("Standard error: {}".format(std_err_brady_fans))

df_brady_fans.describe()
```
