### expenditure is the annual spending on recreation in US dollars,

## Interpret the estimated coefficients. What additional statistics should be given in order to make sure that your interpretations make sense statistically. Write up your answer.

__The t-statistics (and more importantly, their associated p-values) would help us understand the statistical significance of each feature.__

Assuming the p-values are low enough (the lesson suggests < .1 to be statistically significant, and ideally < .05):
* __have_kids__ is a dummy variable, so having kids adds a cost of \\$223.57 per year to a family's annual expenditure.
* __annual_income__ is continuous. I believe this is saying that .0012 of every dollar of income earned is spent on recreation...
* And an additional .00002 of every dollar of income earned, squared, is added to the total spent on recreation. This adds a slight upward bend to the curve as income increases, shown below:

```python
inc_list = []
y = []
have_kids = 1
for income in range(50000,100000,1000):
    inc_list.append(income)
    y.append( 873 + (.0012 * income) + (.00002 * income**2) - (223.57 * have_kids))
    
plt.plot(inc_list,y)
```
